```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import classification_report,accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB

# I have implemented certain algorithms 
```

```python
fake_df=pd.read_csv('/kaggle/input/fake-news-detection/fake.csv')
true_df=pd.read_csv('/kaggle/input/fake-news-detection/true.csv')

# Add Label to Distinguish Between Real and Fake news 
fake_df['label']=0
true_df['label']=1
```

```python
# Combines the Both Dataset in one 
combined =pd.concat([fake_df,true_df],axis=0)

```

```python
# Shuffle the Fake news with Real or True News

combined =combined.sample(frac=1,random_state=42).reset_index(drop =True)
```

```python
combined.isnull().sum()


# It has no null Data
```

```python
combined['content']=combined['title']+" "+combined['text']

combined.drop(['subject','title','date','text'],axis=1,inplace =True)
```

```python
# Initializing the TF-IDF vectorizer 
# It converts the text data into numerical data by assigning importance to words


tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_df=0.7)
X = tfidf_vectorizer.fit_transform(combined['content'])
```

```python
y = combined['label']
```

**Split the Data in Training and Testing**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
print(f"Training Data Shape: {X_train.shape}")
print(f"Testing Data Shape: {X_test.shape}")
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
# Check accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
# Classification report
from sklearn.metrics import confusion_matrix
print(classification_report(y_test, y_pred))

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print(f"Confusion Matrix:\n{conf_matrix}")
```

```python
import pickle

# Save the model
with open('fake_news_model.pkl', 'wb') as model_file:
    pickle.dump(model, model_file)

# Save the vectorizer
with open('tfidf_vectorizer.pkl', 'wb') as vec_file:
    pickle.dump(tfidf_vectorizer, vec_file)
```
